```python
data = train_set.drop("Outcome", axis=1)
data_labels = train_set["Outcome"].copy()
```

```python
data.head()
```

```python
data_labels.head()
```

# Selecting a desired model for Diabetes Prediction

# 1. Using the Logistinc Regression model

```python
# Building a Logistic Regression Model using skicit learn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(data, data_labels)

# predict the target on the train dataset
predict_train = model.predict(data)
print('Target on train data',predict_train) 

# Accuray Score on train dataset
accuracy_train = accuracy_score(data_labels,predict_train)
print('accuracy_score on train dataset : ', accuracy_train)
```

# 2. Using the DecisionTree Classifier model

```python
# Building a DecisionTree Classifier Model using skicit learn
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model2 = DecisionTreeClassifier()
model2.fit(data, data_labels)

# depth of the decision tree
print('Depth of the Decision Tree :', model2.get_depth())

# predict the target on the train dataset
predict_train2 = model2.predict(data)
print('Target on train data',predict_train2) 

# Accuray Score on train dataset
accuracy_train2 = accuracy_score(data_labels, predict_train2)
print('accuracy_score on train dataset : ', accuracy_train2)
```

# 3. Using the RandomForest Classifier model

```python
# Building a RandomForest Classifier Model using skicit learn
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model3 = RandomForestClassifier()
model3.fit(data, data_labels)

# number of trees used
print('Number of Trees used : ', model3.n_estimators)

# predict the target on the train dataset
predict_train3 = model3.predict(data)
print('\nTarget on train data',predict_train3) 

# Accuray Score on train dataset
accuracy_train3 = accuracy_score(data_labels,predict_train3)
print('\naccuracy_score on train dataset : ', accuracy_train3)
```

# Selecting the DecisionTree Classifier model for further prediction. Let's Evaluate the model.

```python
import numpy as np
from sklearn.metrics import mean_squared_error
data_predictions = model3.predict(data)
mse = mean_squared_error(data_labels, data_predictions)
rmse = np.sqrt(mse)
```

# Testing the model on the test dataset.

```python
x_test = test_set.drop("Outcome", axis=1)
y_test = test_set["Outcome"].copy()
```

##### Evaluating the model 

```python
from sklearn.metrics import mean_squared_error
data_predictions = model3.predict(x_test)
mse = mean_squared_error(y_test, data_predictions)
rmse = np.sqrt(mse)
```
